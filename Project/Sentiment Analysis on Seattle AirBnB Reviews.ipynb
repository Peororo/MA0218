## FIlter language

```python
# language
from langdetect import detect
detect(reviews.iloc[-3,-1])
```

```python
reviews[reviews.comments == '']
```

```python
reviews[reviews.comments.isna()]
```

```python
reviews.comments.replace(r'^\s*$', np.nan, regex=True, inplace=True)
reviews.dropna(inplace = True, subset=['comments'])
```

```python
lang_test = reviews.comments.to_list()

for i, rev in enumerate(lang_test):
    try:
        detect(rev)
    except:
        print(i, rev)
```

```python
reviews['language'] = reviews.comments.apply(detect)
```

```python
reviews[reviews.language != 'en']
```

```python
cleaned = reviews[reviews.language == 'en'].reset_index(drop=True)
cleaned.to_csv('AirBnB/clean_reviews.csv')
```

## tokenise

```python
cleaned = pd.read_csv('AirBnB/clean_reviews.csv').drop('Unnamed: 0', axis='columns')
cleaned
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
reviews_list = cleaned.comments.tolist()
reviews_tokenized = [word_tokenize(str(i)) for i in reviews_list[:2]]
reviews_tokenized[0]
```

## porter stemming

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

stemmed = [[stemmer.stem(x) for x in i] for i in reviews_tokenized[:2]]
stemmed[0]
```

## SentimentIntensityAnalyzer

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
for sentence in reviews['comments'].values[:5]:
    print(sentence)
    ss = sid.polarity_scores(sentence)
    for k in sorted(ss):
        print(f'{k}: {ss[k]}')
    print()
```

```python
reviews.comments[:5].apply(sid.polarity_scores)[0]
```

```python
reviews['sentiment'] = reviews.comments.apply(sid.polarity_scores)
```

```python
cleaned = pd.read_csv('AirBnB/clean_reviews.csv').drop('Unnamed: 0', axis = 'columns')
cleaned
```

```python
clean_sent = cleaned.sentiment.str.replace(r"[{}]",'',).str.split(',', expand=True)
clean_sent
```

```python
for col in clean_sent.columns:
    clean_sent[col] = clean_sent[col].str.split(':', expand = True)[1]
    clean_sent[col] = clean_sent[col].astype('float')

clean_sent.columns = ['neg','neu','pos','compound']
clean_sent
```

```python
clean_sent.dtypes
```

```python
sentiment_df = pd.concat([cleaned, clean_sent], axis = 'columns')
sentiment_df.drop('sentiment', axis = 'columns', inplace=True)
sentiment_df
```
